```python
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression

```

```python
df = sns.load_dataset("iris")
df
```

```python
df["species"].unique()
```

```python
df.isnull().sum()
df.info()
df.describe()
df.shape
df.size
```

```python
df1 = df[df['species']!="setosa"]
print(df1.shape)
df
```

```python
df1["species"]=df1["species"].map({"versicolor":0,"virginica":1})
df1
```

```python
x = df1[['petal_width']].values
y= df1['species'].values
```

```python
x_train,x_test, y_train,y_test =train_test_split(x,y,test_size=0.2,random_state= 42)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
x_range = np.linspace(x.min(),x.max(),300).reshape(-1,1)
x_range
```

```python
y_prob = model.predict_proba(x_range)[:,1]
y_prob
```

```python
                                                           plt.figure(figsize=(8, 6))
plt.scatter(x,y,c=y, cmap='coolwarm',edgecolor='k')
plt.plot(x_range,y_prob,color='black',linewidth=2)
plt.xlabel('Petal Width')
plt.ylabel('Probability of Virginica')
plt.title('Logistic Regression Curve for Petal Width')
plt.show()                 
```
